<h1>Реализация логистической регрессии для задачи классификации трех класссов</h1>

<b>Импортирование библиотек</b>

```python
# помимо стандартных библиотек мы также импортируем библиотеку warnings
# она позволит скрыть предупреждения об ошибках
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sqlite3 as sql

# чтобы разделить выборку на обучающую и тестовую с перемешанными данными
from sklearn.utils import shuffle

# для расчета метрики accuracy
from sklearn.metrics import accuracy_score
```

<b>Для начала обработаем базу данных, записав все в датафрейм из pandas, а также посмотрим, сколько классов и как они распределены</b>

```python
# открываем таблицу и записываем все оттуда в датафрейм
with sql.connect('dataset.db') as con:
    cur = con.cursor()
    df = pd.read_sql("SELECT * FROM wine", con)

# убираем столбец индексов, т.к. в датафрейме уже есть свои индексы
df.drop('index', axis= 1 , inplace= True)

# приведем признаки к одному масштабу (отнимаем мат. ожидание и делим на отклонение для каждого столбца)
df = (df - df.mean())/df.std()
 
# добавим целевую переменную заново, тк ее мы тоже привели к одному масштабу, что не надо
df['target'] = pd.read_sql("SELECT target FROM wine", con)
 
# посмотрим, сколько у нас классов и их наблюдений
df.target.value_counts()
```

<b>Перемешаем датафрейм, чтобы разделить его на обучающую и тестовую выборки, не теряя общности</b>

```python
df = shuffle(df)
df
```

<b>Смотрим корреляции признаков и таргет переменных, строя цветовую таблицу</b>

```python
from CorrelationRatio import correlation_ratio

# создадим два списка, один для названий признаков, второй для значений корреляционного отношения
columns, correlations = [], []
 
# пройдемся по всем столбцам датафрейма кроме целевой переменной
for col in df.drop('target', axis = 1).columns:
  # поместим название признака в список columns
  columns.append(col)
  # рассчитаем взаимосвязь этого признака с целевой переменной
  # и поместим результат в список значений корреляционного отношения
  correlations.append(correlation_ratio(df[col], df['target']))
 
# создадим датафрейм на основе заполненных списков
# и применим градиентную цветовую схему
pd.DataFrame({'column': columns, 'correlation': correlations}).style.background_gradient()
```

<b>Можем заметить, что у признаков flavanoids и proline наибольшие корреляции с таргет переменными, посмотрим их на графике</b>

```python
sns.scatterplot(x = df.flavanoids, y = df.proline, hue = df.target, palette = 'bright', s = 50)
plt.show()
```

<b>На графике видно, что можно построить логистическую регрессию, разделив все классы -> берём эти 2 признака для классификации</b>

```python
X = df[['flavanoids', 'proline']]
y = df['target']

# разделяем выборку на обучающую и тестовую
df_copy = df.iloc[120:]
X_learn = X.iloc[0:120]
Y_learn = y.iloc[0:120]

X_test = X.iloc[120:]
Y_test = y.iloc[120:]
```

<b>Импортируем собственноручный класс логистической регрессии и обучаем модель на выборке</b>

```python
from LogReg import SoftmaxLogReg
 
model = SoftmaxLogReg()
 
model.fit(X_learn, Y_learn)
```

<b>Посмотрим на полученные в ходе обучения веса логистической регрессии для каждого из классов</b>

```python
model.thetas
```

<b>Выведем функцию ошибок, чтобы посмотреть, адекватно ли прошла её минимизация градиентным спуском</b>

```python
plt.plot(model.loss)
plt.show()
```

<b>Сделаем прогноз на тестовой выборке и посмотрим на точность предсказаний</b>

```python
y_pred = model.predict(X_test)

accuracy_score(Y_test, y_pred)
```

<b>Выведем графики и пронаблюдаем, где наша модель ошиблась на тестовой выборке</b>

```python
predictions = X_test.copy()
predictions['target'] = y_pred

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (14, 6), gridspec_kw = {'wspace': 0.2, 'hspace': 0.08})
 
sns.scatterplot(data = df_copy, x = 'flavanoids', y = 'proline', hue = 'target',
                palette = 'bright', s = 50, ax = ax1)
 
ax1.set_title('Фактические классы', fontsize = 14)
 
sns.scatterplot(data = predictions, x = 'flavanoids', y = 'proline', hue = 'target',
                palette = 'bright', s = 50, ax = ax2)
 
ax2.set_title('Прогноз SoftmaxLogReg', fontsize = 14)
 
plt.show()
```
